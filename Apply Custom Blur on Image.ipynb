# Blur (Gaussian) an Image using convole2d

#### Load Dependencies

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy.signal import convolve2d
```

#### Load Image

```python
img = mpimg.imread("lena.png")
plt.imshow(img)
```

#### Convert the image into Black & White (take the mean over 2nd axis)

```python
bw_img = img.mean(axis=2)
plt.imshow(bw_img, cmap='gray')
```

#### Making of Gaussian FIlter

```python
blur_filter = np.zeros((20, 20))
for i in range(20):
    for j in range(20):
        distanceFromCenter = (i - 9.5)**2 + (j - 9.5)**2
        blur_filter[i, j] = np.exp(-distanceFromCenter / 50.)

plt.imshow(blur_filter, cmap='gray')
```

#### Apply Convolution on the image with Gaussian Filter to get the blurred picture

```python
out_image = convolve2d(bw_img, blur_filter)
plt.imshow(out_image, cmap='gray')
```

# Blur (Mean/ Normal) an Image using convole2d

#### Making of Mean/ Normal Blur FIlter

```python
mean_filter = np.ones((20, 20))
plt.imshow(mean_filter, cmap='gray')
```

#### Apply Convolution on the image with Mean Blur Filter to get the average blurred picture

```python
out_image = convolve2d(bw_img, mean_filter)
plt.imshow(out_image, cmap='gray')
```
