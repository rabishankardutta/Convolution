# Edge Detection Filter and Convolution

#### Load Dependencies

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy.signal import convolve2d
```

#### Load image

```python
img = mpimg.imread("lena.png")
plt.imshow(img)
```

#### Convert the image into Black & White (take the mean over 2nd axis)

```python
bw_img = img.mean(axis=2)
plt.imshow(bw_img, cmap='gray')
```

#### Making of Edge Detectors Along x & y direction

```python
Hx = np.array([
    [-1,0,1],
    [-2,0,2],
    [-1,0,1],
], dtype=np.float32)

Hy = Hx.T
```

#### Horizontal Edge Detector

```python
convolved_Hx = convolve2d(bw_img, Hx)
plt.imshow(convolved_Hx, cmap='gray')
```

#### Vertical Edge Detector

```python
convolved_Hy = convolve2d(bw_img, Hy)
plt.imshow(convolved_Hy, cmap='gray')
```

#### Final detection

```python
convolved_Final = np.sqrt(convolved_Hx**2, convolved_Hy**2)
plt.imshow(convolved_Final, cmap='gray')
```

#### Gradient Direction

```python
direction_Theta = np.arctan2(convolved_Hy, convolved_Hx)
plt.imshow(direction_Theta, cmap='gray')
```
